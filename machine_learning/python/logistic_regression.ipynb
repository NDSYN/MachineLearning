## Import Libraries

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Data

```python
# import dataset
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset.head()
```

## Select Data

```python
# [rows,columns]
# [:,:-1] - all rows + all columns except the last column

X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

## Split Data (training, test)

```python
# split data: training and test 

from sklearn.model_selection import train_test_split

# random_state = 1; get the same train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

# apply feature scaling on full X data set
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Train Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(X_train,y_train)
```

## Predict New Result

```python
# must transform values to match model scale
classifier.predict(sc.transform([[30,87000]]))
```

```python
classifier.predict([X_test[0]])
```

## Predict Test Set Results

```python
y_pred = classifier.predict(X_test)

# control number of decimals
#np.set_printoptions(precision=2)

# print prediction results side by side
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm = confusion_matrix(y_test,y_pred)
print(cm)

ac = accuracy_score(y_test,y_pred)
print(ac)
```
